```python
dist_freq_quali = pd.DataFrame()
dist_freq_quali_t = list()

for column in df_varquali:
  freq_var = df_varquali[column].value_counts().sort_index()
  perc_var = df_varquali[column].value_counts(normalize=True).sort_index() * 100
  dist_freq_quali_ = pd.DataFrame({'n': freq_var,
                                   '(%)': perc_var})
  dist_freq_quali_t_ = dist_freq_quali_['n'].sum()

  dist_freq_quali = pd.concat([dist_freq_quali, dist_freq_quali_])
  dist_freq_quali_t.append(dist_freq_quali_t_)

dist_freq_quali.to_excel('/content/drive/MyDrive/MBA_TCC/RESULTADOS/Desc_2_2020.xlsx')
```

```python
corr1 = selcol.corr(method='pearson')
print(corr1)
corr2 = selcol.corr(method='spearman')
print(corr2)

plt.imshow(corr2, cmap='hot', interpolation='none')
plt.colorbar()
plt.title('2020')
plt.xticks(range(len(corr2)), rotation=90, labels=['CN','CH','LC','MT','Redação'])
plt.yticks(range(len(corr2)), rotation=0, labels=['CN','CH','LC','MT','Redação']);
plt.show()
```

```python
plt.figure(figsize=(15,8))
matriz = np.triu(corr2)
sns.heatmap(corr2, mask= matriz, annot=True, cmap='YlGnBu')
plt.title('2020')
plt.xticks(range(len(corr2)), rotation=0, labels=['CN','CH','LC','MT','Redação'])
plt.yticks(range(len(corr2)), rotation=0, labels=['CN','CH','LC','MT','Redação']);
plt.savefig("/content/drive/MyDrive/MBA_TCC/RESULTADOS/Notas_2020 - Matriz de Correlacao (S).png")
plt.show()
```
